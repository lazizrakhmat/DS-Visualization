```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.info()
```

## Chiziqli grafig chizamiz

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.plot(df)
plt.show()
```

## Keling endi Grafigimizni Epaqaga keltiramiz😁

##! Chiqizli Grafiglardan biz nimani ko'rishimiz mumkin asosiy maqsadi `tendensiya` ni ko'rish ya'ni vaqt o'tishi bilan biror bir jarayonni `qanday o'zgarishi`ni ko'rishdir.

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashriflar soni")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(data=df)
plt.show()
```
